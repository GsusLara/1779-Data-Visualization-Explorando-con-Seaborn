```python
import pandas as pd
```

```python
datos = pd.read_csv('credit_card.csv')
```

```python
datos.head()
```

```python
datos.info()
```

###Traduciendo los titulos

```python
datos.columns
```

```python
cabecera = {
    'LIMIT_BAL' : 'Limite',
    'CHECKING_ACCOUNT':'cuenta_corriente',
    'EDUCATION':'escolaridad',
    'MARRIAGE':'estado_civil',
    'AGE':'edad',
    'BILL_AMT':'valor_de_factura',
    'PAY_AMT':'valor_de_pago',
    'DEFAULT':'moroso'
}
```

```python
tarjetas = datos.rename(columns = cabecera)
```

```python
tarjetas.head()
```

```python
tarjetas.cuenta_corriente.unique()
```

```python
nombreCc = {
    'Yes':'Si',
    'No':'No'
}
```

```python
tarjetas.cuenta_corriente.map(nombreCc)
```

```python
tarjetas.head()
```

```python
tarjetas.cuenta_corriente = tarjetas.cuenta_corriente.map(nombreCc)
```

```python
tarjetas
```

```python
tarjetas.escolaridad.unique()
```

```python
nombresEscolaridad = {
    '2.University':'Universidad', '3.Graduate School': 'Escuela', '1.High School':'Colegio'
}
```

```python
tarjetas.escolaridad = tarjetas.escolaridad.map(nombresEscolaridad)
```

```python
tarjetas.estado_civil.unique()
```

```python
nombresEstadoC= {'Married': 'Casado', 'Single': 'Soltero' }
```

```python
tarjetas.estado_civil = tarjetas.estado_civil.map(nombresEstadoC)
```

```python
tarjetas
```

#**Analisis de la informacion con seaborn**

```python
import seaborn as sns
```

```python
!pip show seaborn
```

```python
tarjetas.head()
```

```python
sns.distplot(tarjetas['Limite'])
```

```python
sns.displot(data= tarjetas, x = 'Limite', hue = 'escolaridad');
```

```python
tarjetas['indice_de_uso'] = tarjetas['valor_de_factura'] / tarjetas['Limite']
```

```python
tarjetas.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.displot(data = tarjetas, x ='indice_de_uso',hue = 'escolaridad', palette = "inferno");
```

```python
sns.countplot(data=tarjetas, x="cuenta_corriente", hue="moroso");
```

```python
grupos = [20,30,40,50,100]
rangos = ['entre 20 - 30', 'entre 30 - 40', 'entre 40 - 50', 'mayores a 50']
tarjetas['rango_edad'] = pd.cut(tarjetas['edad'], grupos, labels=rangos)
```

```python
tarjetas.head()
```

```python
sns.boxplot(data=tarjetas, x="rango_edad", y="Limite");
```

```python
tarjetas.head()
```

```python
sns.displot(data=tarjetas, x='Limite', col='escolaridad', kind='kde', hue='rango_edad');
```

```python
sns.scatterplot(data=tarjetas, x='indice_de_uso', y='valor_de_factura', hue= 'cuenta_corriente');
```
